```python
from bs4 import BeautifulSoup
import requests
import re
# how we can extract all the links from a URL or an HTML document using Python.

def getHTMLdocument(url):


    response = requests.get(url)
    return response.text



url_to_scrape = "https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"

html_document = getHTMLdocument(url_to_scrape)
soup = BeautifulSoup(html_document, 'html.parser')
for link in soup.find_all('a',
                          attrs={'href': re.compile("^https://")}):
    # display the actual urls
    print(link.get('href'))
```

```python
import requests
from bs4 import BeautifulSoup
#extract all images link

def getdata(url):
    r = requests.get(url)
    return r.text

htmldata = getdata("https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1")
soup = BeautifulSoup(htmldata, 'html.parser')
for item in soup.find_all('img'):
    print(item['src'])
```

```python
import urllib.request
url = "https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"

try:
    response = urllib.request.urlopen(url)
    data = response.read()
    html_content = data.decode('utf-8')
    print(html_content)

except Exception as e:
    print("Error fetching URL:", e)
```
