```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={"date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01"]),
                            "value":[10,15,22,28,35]}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("date")
plt.ylabel("value")
plt.title("time series with trend")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
text_data="Bhumika and Greeshma are super Monkeys"
wordcloud=WordCloud(width=800,height=400,background_color='black').generate(text_data)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title('word cloud')
plt.axis('off')
plt.show()
words=text_data.split()
word_counts=Counter(words)
plt.figure(figsize=(10,5))
plt.bar(word_counts.keys(),word_counts.values())
plt.title('histogram of words frequency')
plt.xlabel('words')
plt.ylabel('frequency')
plt.xticks(rotation=90)
plt.show
```
